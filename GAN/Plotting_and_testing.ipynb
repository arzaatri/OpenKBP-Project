```python
import torch
import os

def read_items(path):
    for item in os.listdir(path):
        if item == '.ipynb_checkpoints':
            continue
        item_path = f'{path}/{item}'
        item_name = item.split(".")[0]
        exec(f'global {item_name}; {item_name} = torch.load(item_path, map_location="cpu")')
```

```python
import torch
from dataloaders import get_dataloaders
from pix2pix import get_models
import train
import os
from doseloss import DoseLoss

def score_test(generator, dataloader):
    doseloss = DoseLoss()
    loss = 0
    batch_size = dataloader.batch_size
    for batch in dataloader:
        with torch.no_grad():
            true_A, true_B, mask = batch
            fake_B = generator(true_A)
            dl = doseloss(fake_B, true_B, mask)
            loss += dl.item()*batch_size
    loss = loss / dataloader.dataset.__len__()
    return loss

datasets, dataloaders = get_dataloaders(16,16,16)
discriminator, generator = get_models(4,2,16,4, gen_type=2)

loss_dict = {}
for item in os.listdir('saved_items'):
    if item in ['.ipynb_checkpoints', 'untitled.txt', '0.001-1-1.0-old']:
        continue
    path = 'saved_items'+item
    g = torch.load(f'saved_items/{item}/generator_dict.pt', map_location=torch.device('cpu'))
    generator.load_state_dict(g)
    loss_dict[item] = score_test(generator, dataloaders['test'])
    print(f'Loss for {item}: {loss_dict[item]}')
        
torch.save(loss_dict, 'saved_items/loss_dict.pt')
```
